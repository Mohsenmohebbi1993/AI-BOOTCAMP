> 1. Type Conversion and Working with Inputs  
Write a program that takes the user's name and age, then prints a sentence in the following format:  
Hello (name), you are (age) years old

```python
Name = input("Enter Your Name")
Age = int(input("Enter Your Age"))
print("Hello %s , you are %d years old" % (Name, Age))
print(f"Hello {Name} , you are {Age} years old")
```

> 2. Working with Strings (String Indexing)  
Take a string from the user and print its first and last letters.

```python
Text = input("Enter one world: ")
first , end= Text[0], Text[-1]
print(f"World is {Text} , Fisrt is {first} and End is {end}")
```

3. List Management  
Create a list containing five numbers, then: 
- Print the third number 
- Change the second number 
- Print the entire list after the modification 

```python
My_list = [2, 5, 9, 8, 10]
print(f"list is                        {My_list}")
print(f"Size of list is                {len(My_list)}")
print(f"3rd Number in list is          {My_list[2]}")

My_list[1] = 100
print(f"New list is(change 2nd Number) {My_list}")

```
