> 7. Calculate the Average of Grades  
Have a list of grades (e.g., `[12, 18, 15, 20, 10]`), calculate their average, and if the average is greater than 10, print "Passed"; otherwise, print "Failed".

```python
Grades = [12, 18, 15, 20, 10]

Avrage_grades = sum(Grades) / len(Grades)

if Avrage_grades > 10:
    print(f"Avrage Grades is {Avrage_grades}, Passed")
else:
    print(f"Avrage Grades is {Avrage_grades}, Failed")
```

> 8. Vowels and Consonants  
Take a string and use a `for` loop to count and print the number of vowels (a, e, i, o, u) in it.

```python
Sentence = ("Life is not a rehearsal.\nEach day is a new show, no repeat no rewind.\nPerform carefully live the best, choose the best and do the best")
print(Sentence)

Vowels = ["a", "e", "i", "o", "u"]

Count_Vowels = 0
for i in range(len(Sentence)):

    if Sentence[i] in Vowels:
        Count_Vowels += 1


print(f"--- Count Vowels = {Count_Vowels} ---")
```

> 9. Write a program that takes numbers from the user until zero is entered, then prints the sum of all entered numbers at the end.

```python
List_of_number=[]
Number = int(input("Enter Number"))

while Number != 0:
    List_of_number.append(Number)
    Number = int(input("Enter Number"))


print("list of number is %d\n And Sum of All number is %s" % (sum(List_of_number), List_of_number))

```

> 10. Student List Management

> 1. **Student grades**  

- Create a list of student grades.  
- Calculate the average, and find the minimum and maximum grades.  
- Store the grades that are higher than the average in a separate list.  

```python
print("---- Student Grades ----")

Grades = [10, 20, 20, 18, 8, 2, 17, 13, 20, 10, 12, 8, 9, 3, 1, 0, 0, 18, 16, 17, 19]
Count_students = len(Grades)
Avrage_grades = sum(Grades) / Count_students
min_grades = min(Grades)
Max_grades = max(Grades)
Grades_higher_than_average = []
for i in Grades:
    if i > Avrage_grades:
        Grades_higher_than_average.append(i)

print(f"Number of students: {Count_students}\nAvrage Grades: {Avrage_grades}")
print(f"min Grades: {min_grades}\tMax Grades: {Max_grades}")
print(f"List of Grades Higher than Avrage:\n{Grades_higher_than_average}")
```
