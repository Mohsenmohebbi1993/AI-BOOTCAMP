# Numpy

NumPy (Numerical Python) is one of the most important Python libraries for performing numerical computations and working with arrays and multidimensional matrices. The core of this library is a data structure called ndarray, which enables fast storage and processing of numerical data. NumPy performs many mathematical, statistical, and matrix operations efficiently and with high speed (because it is implemented in the C language), and for this reason it is considered the foundation of many scientific and data‑oriented Python libraries.

Applications of NumPy:

* Processing and analyzing numerical data  
* Performing matrix operations and linear algebra  
* Use in Data Science  
* Implementing and training Machine Learning models  
* Image and signal processing  
* Performing simulations and scientific computations  
* The foundation for libraries such as Pandas, SciPy, TensorFlow, and PyTorch

### Universal Function in numpy

```python
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output
    
values = np.random.randint(1, 10, size=5)
compute_reciprocals(values)
```

```python
big_array = np.random.randint(1, 100, size=1_000_000)
%timeit compute_reciprocals(big_array)
```

```python
# big_array is one array and can use numpy
%timeit (1.0 / big_array)
```

> NumPy Operators — Cheat Sheet  

---

**Arithmetic Operators**

| Operator | NumPy Function | Description |
|---------|----------------|-------------|
| `+` | `np.add(a, b)` | Element‑wise addition |
| `-` | `np.subtract(a, b)` | Element‑wise subtraction |
| `*` | `np.multiply(a, b)` | Element‑wise multiplication |
| `/` | `np.divide(a, b)` | Element‑wise division |
| `//` | `np.floor_divide(a, b)` | Floor (integer) division |
| `%` | `np.mod(a, b)` / `np.remainder(a, b)` | Remainder |
| `**` | `np.power(a, b)` | Exponentiation |
| `-a` | `np.negative(a)` | Negation |

---

**Comparison Operators**

| Operator | NumPy Function | Description |
|---------|----------------|-------------|
| `==` | `np.equal(a, b)` | Element‑wise equality |
| `!=` | `np.not_equal(a, b)` | Element‑wise inequality |
| `<` | `np.less(a, b)` | Less than |
| `<=` | `np.less_equal(a, b)` | Less than or equal |
| `>` | `np.greater(a, b)` | Greater than |
| `>=` | `np.greater_equal(a, b)` | Greater than or equal |

---

**Logical Operators**

| Operator | NumPy Function | Description |
|---------|----------------|-------------|
| `&` | `np.logical_and(a, b)` | Logical AND |
| `\|` | `np.logical_or(a, b)` | Logical OR |
| `~` | `np.logical_not(a)` | Logical NOT |
| `^` | `np.logical_xor(a, b)` | Logical XOR |

**Important:** Always use parentheses with logical operators in NumPy:  
Example: `(a > 0) & (b < 5)`

---

**Bitwise Operators**

| Operator | NumPy Function | Description |

| `^` | `np.bitwise_xor(a, b)` | Bitwise XOR |
| `~` | `np.invert(a)` | Bitwise NOT |
| `<<` | `np.left_shift(a, b)` | Left shift |

```python
days = np.arange(365)

MASK_RAIN = (inches > 0)
MASK_SUMMER = (days > 172) & (days < 262)

print(f"Median rain in 2024 {np.median(inches[MASK_RAIN]):.4f}")
print(f"Median rain in semmer 2024 {np.median(inches[MASK_SUMMER]):.4f}")
print(f"Max rain in semmer 2024 {np.max(inches[MASK_SUMMER]):.4f}")
print(f"days rain in non-semmer 2024 {np.count_nonzero(inches[MASK_RAIN & ~MASK_SUMMER])}")
```
