Basic method
1. lower  
2. upper    
3. strip  
4. replace  
5. split  
6. join  

```python
text = "I am mohsen Mohebbi"
print(f"main text = {text}, lower text = {text.lower()}")
print(f"main text = {text}, upper text = {text.upper()}")
print(f"main text = {text}, titel text = {text.title()}")

# have a lot of space First an End
text = "     I am mohsen    Mohebbi     "
print(f"main text = {text}, strip text = {text.strip()}|") # strip as 2 said
print(f"main text = {text}, rstrip text = {text.rstrip()}|") # strip as right
print(f"main text = {text}, lstrip text = {text.lstrip()}|") # strip as left

text = "------I am mohsen---Mohebbi------"
print(f"main text = {text}, strip text = {text.strip("-")}|") # strip as 2 said on "-"

text = "---   I am mohsen---Mohebbi    ----"
print(f"main text = {text}, strip text = {text.strip("-").strip()}|")

# Replce method
text = "I am mohsen Mohebbi"
print(f"main text = {text}, Replce = {text.replace("mohsen","hasan")}")

text = "  I am mohsen Mohebbi  "
print(f"main text = {text}, Replce = {text.replace("  ","--")}")

# split method
text = "I am mohsen Mohebbi"
print(f"main text = {text}, split = {text.split()}")

# Find method
text = "I am mohsen Mohebbi"
print(f"main text = {text}, find = {text.find("m")}") # result -1: cant find or there is not


# Join method
text = "I am mohsen Mohebbi"
text = text.split()
print(f"main text = {text}, join = {'-'.join(text)}")
```

```python
name = "mohsen"
last_name = "mohebbi"
age = 33

user_data = [name, last_name, age]
print(user_data)

# append 
user_data.append(70.8)  # .append(value) -> add value at end
print(user_data)

# insert
user_data.insert(0, 10000)  #.insert(loc, value)
print(user_data)

# pop
print(f"pop  is {user_data.pop()}")  #.pop(index) remove last value (difalt is last)
print(user_data)

# remove -> first item that is equal remove-method , remove
# value should in list
print(f"remove  is {user_data.remove(10000)}")  #.remove(value) ->None: have not result
print(user_data)

user_data.append(200)
user_data.append(200)
print(user_data)
print(f"remove  first 200 {user_data.remove(200)}")
print(user_data)

# count
print(f"count Number 200  = {user_data.count(200)}")

# index
print(f"index of 33 is = {user_data.index(33)}")
```

> ## Project 2  
 
1. Create one String from Number   
2. find : `Len`, `min`, `Max`, `Avg`, `Sum`, `Sort`, `Uniq`  
3. Print by `f-string`

```python
# Creat one String 
string = "1, 2, 3,8,9,  10,1 ,23, 12,      1, 23     , 19"

# Clear : remove space
strint_of_clear = string.replace(" ","")

# Slicing and Creat one list 
strint_of_clear = strint_of_clear.split(",")

# change string to int
strint_of_clear_int = [int(i) for i in strint_of_clear]

# sort

# Print
print(f"Len : {len(strint_of_clear_int)}",end="")
print(f"\t Max : {max(strint_of_clear_int)}",end="")
print(f"\t min : {min(strint_of_clear_int)}",end="")
print(f"\t Sum : {sum(strint_of_clear_int)}",end="")

print(f"\t Avg : {(sum(strint_of_clear_int) / len(strint_of_clear_int)):.2f}")
print(f"List: {strint_of_clear_int}")

# sort
strint_of_clear_int.sort()
print(f"Sort: {strint_of_clear_int}")

# Uniq
print(f"Uniq: {list(set(strint_of_clear_int))}")



```
