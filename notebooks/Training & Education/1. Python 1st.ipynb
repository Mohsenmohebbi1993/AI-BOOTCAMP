> ## Guessing Game  


Play by `for` and `if`

Save a Number  
Input a number and say "is this namber True of False  
Help to find Number  

> Example 1: Guessing in a Range from **1** to **100** 


Suppose the user defines the range from **1** to **100**, and the system randomly selects **42** as the target number. The guessing process might look like this:
* Guess 1: 50 → Too high  
* Guess 2: 25 → Too low  
* Guess 3: 37 → Too low  
* Guess 4: 43 → Too high  
* Guess 5: 40 → Too low  
* Guess 6: 41 → Too low  
* Guess 7: 42 → Correct!  
*Total Guesses: 7*

> Algorithm  


Accept lower and upper bounds from the user.  
Generate a random number in the selected range.  
Calculate the maximum allowed guesses using the binary search formula.  
Run a loop to take user guesses:  
If the guess is too high, print: _*"Try Again! You guessed too high."*_  
If the guess is too low, print: _*"Try Again! You guessed too small."*_    
If the guess is correct, print: _*"Congratulations!" and exit the loop.*_    
If the user runs out of chances, display the correct number and a message: _*"Better Luck Next Time!"*_  


````python
BASE_NUMBER = 32
Count_respite_loop = 0
while True:
    Count_respite_loop += 1
    user_input = int(input("Enter Your Number:  ")) # give input from `USER`
    if user_input == BASE_NUMBER:
        print("-Congratulations! and exit the loop")
        break
    elif user_input > BASE_NUMBER:
        print("-Try Again! You guessed too high")
    else:
        print("-Try Again! You guessed too small.")

print(f"Total Guesses: {Count_respite_loop} - Number is {user_input}")

````

> ## Check Prime Number in Python  

Given a positive integer **N**, the task is to write a Python program to check if the number is Prime or not in Python.  
For example, given a number **29**, it has no divisors other than **1** and **29** itself. Hence, it is a prime number.  



````python
user_input = int(input("Enter Your number: "))
is_preime = True

for i in range(  2 ,   int((user_input **0.5)+1)   ):
    if not user_input % i:
        is_preime = False
        break

if is_preime:
    print(f"Your Number {user_input} is Peime")
else:
    print(f"Your Number {user_input} is NOT Peime")
````

Find Numbers of `0` to `100`, that are Preime

````python
preime_number_list=[]


for i in range (2, 101):
    is_preime = True
    for j in range(2, int((i**0.5)+1)):
        if not i % j:
            is_preime = False
            break
    if is_preime:
        preime_number_list.append(i)


print(preime_number_list)
````

````python
def calculator(first_number: float, second_number: float, opration_type: str = "sum"):

    # Make Opration Type Clear
    opration_type = opration_type.lower() # all Charechter lower
    opration_type = opration_type.strip() # Remove Space in first and end

    # COndition Section
    if opration_type == "sum":
        value = first_number + second_number
        print(f"-first_number + second_number = {value}")
    elif opration_type == "sub":
        value = first_number - second_number
        print(f"-first_number - second_number = {value}")
    elif opration_type == "devision":
        value = first_number / second_number
        print(f"-first_number / second_number = {value}")
    elif opration_type == "pow":
        value = first_number ** second_number
        print(f"-first_number ** second_number = {value}")
    else:
        print(f"-Please Enter a valid value as Opration Type")


# Run function Of calculator by a Dinamic menu
while True:
    print("Enter Your Number and Operation:")
    print("1. sum\n2. sub\n3. devision\n4. pow\n0. END")
    opration_type = (input("opration_type:  "))
    if opration_type == "end":
        break
    first_number, second_number = float(input("first_number:  ")), float(input("second_number:  ")) 
    calculator(first_number, second_number, opration_type)
    
````

A **lambda function** is a one‑line function that takes **one input** and returns **one output**.  
It must have at least **one input argument**.

Lambda functions are mainly useful when working with other functions that **accept a function as an input**, especially

````python
my_function = lambda x: x**2

print(my_function(8))
````

````python
my_list = list(range(20))

# Reteurn Even Number

# first code by for
Even_list = []
for i in my_list:
    if not i % 2:
        Even_list.append(i) 
print(f"Return first code {Even_list}")

# Second code by lambda
Even_list = list(filter(lambda number: not number % 2 , my_list)) # filter( function, list)
print(f"Return second code {Even_list}")
````

Student Grades and Average Calculation

We have a list of students' scores, and we want to assign them grades **A, B, C, or D**.  
After that, we calculate the **average score** and display it to the user.


````python
scores = [82, 92, 67, 43, 74, 59]

def get_grade(scores:list):
    if scores >= 90:
        return  "A"
    elif scores >= 80:
        return  "B"
    elif scores >= 70:
        return  "C"
    elif scores >= 60:
        return  "D"
    else:
        return "F"

gardes = {}

for i, score in enumerate(scores):
    gardes[f"student_{i+1}"] = get_grade(score)

Avrage_scores = sum(scores) / len(scores)

print("Grades:")
for student, g in gardes.items():
    print(f"{student} : {g}")

print(f"Avrage Scores is {Avrage_scores}")
 

 
````
