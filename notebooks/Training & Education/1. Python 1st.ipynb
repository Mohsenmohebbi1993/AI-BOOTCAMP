Operators in Python

```python
print("Number is 10")
x = 10
print(f"{x} + 2 = {10 + 2}\n{x} - 2 = {10 - 2}\n{x} * 2 = {10 * 2}\n{x} / 2 = {10 / 2}")
print(f"{x} // 4 = {10 // 4}\n{x} % 4 = {10 % 4}")
```

Setting in Print

```python
# print 2 objet has one space together
# end of print has one Enter
print("hello", 100*2 , 500/3, "mohsen")
print("***test***")

# change Space with --   (" " --> "--")
# remove Enter   (\n --> "")
print("hello", 100*2 , 500/3, "mohsen", sep="--", end="")
print("***test***")
```

Sum , Max , Min

```python
print("sum([1, 2, 3] = %d)" % sum([1, 2, 3]))
print("min([1, 2, 3] = %d)" % min([1, 2, 3]))
print("max([1, 2, 3] = %d)" % max([1, 2, 3]))
```

Data Type

```python
user_age: int = 23
print(f"user_age = {user_age}, type = {type(user_age)}")

user_weight: float = 69.75
print(f"user_weight = {user_weight}, type = {type(user_weight)}")

user_name: str = "Mohsen Mohebbi"
print(f"user_name = {user_name}, type = {type(user_name)}")

is_student: bool = True
print(f"is_student = {is_student}, type = {type(is_student)}")
```
