> ## Guessing Game  


Play by `for` and `if`

Save a Number  
Input a number and say "is this namber True of False  
Help to find Number  

> Example 1: Guessing in a Range from **1** to **100** 


Suppose the user defines the range from **1** to **100**, and the system randomly selects **42** as the target number. The guessing process might look like this:
* Guess 1: 50 → Too high  
* Guess 2: 25 → Too low  
* Guess 3: 37 → Too low  
* Guess 4: 43 → Too high  
* Guess 5: 40 → Too low  
* Guess 6: 41 → Too low  
* Guess 7: 42 → Correct!  
*Total Guesses: 7*

> Algorithm  


Accept lower and upper bounds from the user.  
Generate a random number in the selected range.  
Calculate the maximum allowed guesses using the binary search formula.  
Run a loop to take user guesses:  
If the guess is too high, print: _*"Try Again! You guessed too high."*_  
If the guess is too low, print: _*"Try Again! You guessed too small."*_    
If the guess is correct, print: _*"Congratulations!" and exit the loop.*_    
If the user runs out of chances, display the correct number and a message: _*"Better Luck Next Time!"*_  


````python
BASE_NUMBER = 32
Count_respite_loop = 0
while True:
    Count_respite_loop += 1
    user_input = int(input("Enter Your Number:  ")) # give input from `USER`
    if user_input == BASE_NUMBER:
        print("-Congratulations! and exit the loop")
        break
    elif user_input > BASE_NUMBER:
        print("-Try Again! You guessed too high")
    else:
        print("-Try Again! You guessed too small.")

print(f"Total Guesses: {Count_respite_loop} - Number is {user_input}")

````

> ## Check Prime Number in Python  

Given a positive integer **N**, the task is to write a Python program to check if the number is Prime or not in Python.  
For example, given a number **29**, it has no divisors other than **1** and **29** itself. Hence, it is a prime number.  



````python
user_input = int(input("Enter Your number: "))
is_preime = True

for i in range(  2 ,   int((user_input **0.5)+1)   ):
    if not user_input % i:
        is_preime = False
        break

if is_preime:
    print(f"Your Number {user_input} is Peime")
else:
    print(f"Your Number {user_input} is NOT Peime")
````

Find Numbers of `0` to `100`, that are Preime

````python
preime_number_list=[]


for i in range (2, 101):
    is_preime = True
    for j in range(2, int((i**0.5)+1)):
        if not i % j:
            is_preime = False
            break
    if is_preime:
        preime_number_list.append(i)


print(preime_number_list)
````

Avoiding Repeated Code with Inheritance

Whenever you notice that **a piece of code you wrote earlier is being written again**, it means there is another (better) way to structure your program.

For example, suppose the `Person` class stores a person’s **first name, last name, and age**.  
Now, imagine we want to create a new class for **employees**. Although the main purpose of the `Staff` class is not to handle basic personal information, we *know* that every employee still needs to have a first name, last name, and age.

When we create the `Staff` class, we might see that some parts of the `Person` class are being **repeated**, requiring us to write the same code again.

In such a situation, the second class can **inherit** from the main class.  
This allows the child class to automatically receive attributes like **first name** and **last name** from the higher class (the **Base Class** or **Parent Class**), and the new class becomes the **Child Class** or **Derived Class**.


````python

class Person:
    """ Most related attributes to Person """

    def __init__(self, first_name, last_name, age):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

    def get_fullname(self):
        return f"{self.first_name} {self.last_name}"

    def introduce(self):
        return f"I am {self.get_fullname()}, and I am {self.age}"

    def __str__(self):
        """ when call Variable  
        Return one srting
        """
        return self.introduce() 


# --------- Inheritance ---------

class staff(Person): # <----- class name_child_class(name_Parent_class)
    """ Most related attributes to each staff """
    pass


person_1 = Person("David", "Doe", 45)
person_2 = Person("Robert", "Rabinson", 26)

# Without writing additional code,
# it automatically asks for the same information
# required by the `Person` class,
# because it has inherited from the parent class.
staff_1 = staff("John", "Stark", 32)


print(person_1)
print(person_2)

print(staff_1)
print(staff_1.first_name)
````

> Calling the Parent Class's Constructor

When the `Staff` class inherits from the `Person` class, we often want to reuse attributes like `first_name`, `last_name`, and `age` that are already defined in the `Person` class, without redefining them in `Staff`.

Instead of repeating the code, we use the `super()` function.

`super()` essentially means:

**"Refer to my parent (the `Person` class)."**


> Encapsulation in OOP

**Encapsulation** is a concept in Object-Oriented Programming (OOP) that means **bundling data (attributes) and the methods that operate on that data inside a class**, while **restricting direct access to some of the object's internal details**.

The main goal of encapsulation is to **protect the internal state of an object** and control how its data is accessed or modified.

Instead of accessing attributes directly, we usually interact with them through **methods** (such as getters and setters). This helps maintain **data integrity** and keeps the implementation details hidden from outside the class.

> Protecting Data with Encapsulation

If you want to **protect your data** so that a user cannot easily access certain methods or attributes, you can use **encapsulation**.

Encapsulation allows you to:

- **Restrict direct access** to sensitive attributes  
- **Control reading and writing** through getter and setter methods  
- **Limit access** to internal functionality so that users cannot misuse or modify important parts of the code  

This ensures that the object’s data remains safe, consistent, and only changeable in controlled ways.



In Python, when we use `property`, we can define three things for an attribute:

- **getter** → to read the value

> Method Resolution Order (MRO) in Python

**MRO** means:

The order in which Python searches for **methods and attributes** in classes.

When you call a **method** or **attribute**, Python must decide **which class it should look in** to find it.

**MRO determines:**
- Where Python should **start searching**
- And **in what order** it should continue through the classes.

EX"
```python

        print("A")

class B(A):
    pass

    pass

obj = C()  # C → B → A

print(C.mro())
>>> [C, B, A, object]

```

> ### Error Handeling

> * Types of Errors in Python

1. **SyntaxError**  
   Occurs when the Python syntax is incorrect (e.g., an unclosed parenthesis).

2. **IndentationError**  
   Happens due to incorrect indentation (wrong spacing in code blocks).

3. **NameError**  
   Raised when a variable or function is used before being defined.

4. **TypeError**  
   Occurs when an operation is applied to an inappropriate data type (e.g., adding a number to a string).

5. **ValueError**  
   The data type is correct, but the value itself is invalid (e.g., converting a non‑numeric string to `int`).

6. **IndexError**  
   Raised when accessing an index outside the valid range of a list or array.

7. **KeyError**  
   Occurs when trying to access a dictionary key that does not exist.

8. **AttributeError**  
   Happens when an object does not have the requested attribute or method.

9. **ImportError**  
   Raised when a module or a function cannot be imported.

10. **ModuleNotFoundError**  
    A subtype of `ImportError`; occurs when the module itself cannot be found.

11. **ZeroDivisionError**  
    Raised when attempting to divide by zero.

12. **FileNotFoundError**  
    Occurs when trying to open a file that does not exist.

13. **IOError / OSError**  
    Input/output errors, such as problems accessing files or the operating system.

14. **RuntimeError**  
    A general runtime error when no more specific category applies.

15. **MemoryError**  
    Raised when the system runs out of memory for an operation.

16. **OverflowError**  
    Occurs when a mathematical operation returns a value too large to be represented.

> Regular Expressions  

Regular Expressions, or Regex, in Python are a powerful way to search, match, extract, and replace text patterns in strings. Using the `re` module, you can define patterns that detect specific text structures—for example, identifying emails, phone numbers, numbers, or certain words within a large text.  

Instead of searching for a fixed string, Regex uses patterns such as `\d` for digits, `\w` for word characters, and `.` for any character. Then, functions like `re.search()` are used to find the first match, `re.findall()` to find all matches, and `re.sub()` to replace text.  

Regex is widely used in text processing, data validation, information extraction from files, and web data.
