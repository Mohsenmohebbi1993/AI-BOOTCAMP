tensor from start to end by step step      
`torch.arange(start, end, step)`   
result:    [ start, ... , end )

```python
arange_tensor = torch.arange(0, 10, 1)
print(arange_tensor)
```

```python
arange_tensor = torch.arange(1.5, 2.5, 0.1)
print(arange_tensor)
```

tensor between start and end and count of selection   
`torch.linspace(start, end, count of selection)`    
result:   [  start, ... , end  ]

```python
linspace_tensor = torch.linspace(0, 1, 100)
print(linspace_tensor)
```

```python
a = torch.rand(2,2)
b = torch.rand(2,2)
print(a)
print(b)
c = a + b
print(c)
```

```python
a = torch.rand(1, 2, 3)
print(a)
b = 2.5
c = a + b
print(c)
```

out=input+alpha×other

```python
alpha = 0.3
c = a + alpha * b
print(c)
```

```python
# add(input, other, *, alpha=1, out=None) -> Tensor

d = torch.add(a, b, alpha=alpha)
print(d)
```

```python
b1 = torch.randn(4)
c1 = torch.randn(4, 1)
print(b1)
print(c1)
print(b1 + c1 * 10)
print(torch.add(b1, c1, alpha=10))
```

```python
a = torch.randint(10, (4,2))
print(a)
b = torch.randint(10, (2,))
print(b)
c = a * b
print(c)
```

```python
a = torch.randn(4,2)
b = torch.randn(2,4)
print(a)
print("---------------")
print(b)
print("---------------")

c = torch.matmul(a, b)
print(c)
```

```python
c = a @ b
print(c)
```
