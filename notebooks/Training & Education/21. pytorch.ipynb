Zeros in torch

```python
zeros1d = torch.zeros(1)
print(zeros1d)
```

```python
zeros1d = torch.zeros(5)
print(zeros1d)
```

```python
zeros2d = torch.zeros(2, 3)
print(zeros2d)
```

```python
zeros5d = torch.zeros(2, 3, 8, 6, 5)
print(zeros5d.shape)
print(zeros5d)
```

numel : number of element of one tensor

```python
print(zeros5d.shape)
manual_count_element = 1
for i in (zeros5d.shape):
    manual_count_element *= i

print(f"manual count element : {manual_count_element}")

```

```python
print(zeros5d.numel())
```

change type

```python
zeros2d = torch.zeros(2, 2, dtype=torch.float64)
print(zeros2d)
```

```python
ones1d = torch.ones(4)
print(ones1d)
```

```python
ones2d = torch.ones(4, 8)
print(ones2d)
```
