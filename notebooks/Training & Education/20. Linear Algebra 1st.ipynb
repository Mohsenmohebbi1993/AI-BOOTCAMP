## Opetarors in Vectors

Addition and subtraction

```python
s = np.array([20 , 15])
m = np.array([25 , 10])

print(f"s =     {s}")
print(f"m =     {m}")

t_p = s + m
print(f"s + m = {t_p}")

t_p = np.add(s, m)
print(f"s + m = {t_p}")

t_p = np.sum([s, m], axis=0)
print(f"s + m = {t_p}")


t_n = s - m
print(f"s - m = {t_n}")

t_p = np.subtract(s, m)
print(f"s - m = {t_p}")

```

power

```python
a = np.array ([12 , 10])
print(a)

print(a*3)
```

dot product

```python
a = np.array([2 , 4])
b = np.array([4 , 10])

print(a)
print(b)

print(np.dot(a , b))
```

```python
a = np.array([3 , -4 , 5])
b = np.array([-1 , 1/2 , 10])

print(a)
print(b)

print(np.dot(a , b))
```

## Matrices

```python
# create one Matrix
M=np.array([[20,12],
            [25,15],
            [23,13]])
print(M)
```

```python
# Matrix size
print(np.shape(M))
```

```python
M=np.array([[2,6],
            [-1,5]])
print(M)

print(np.shape(M))
```

```python
# diagonal
t=np.diag(M)
print(t)
```

```python
# Matrix in numpy
A=np.asmatrix([[1,0,5],
          [-2,6,7],
          [3,10,15]])
print(A)
```

```python
t=np.diag(A)
print(t)
```
