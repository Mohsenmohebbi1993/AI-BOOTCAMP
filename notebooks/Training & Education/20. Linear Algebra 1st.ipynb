# Linear Algebra

We use `numpy` for Linear Algebra

```python
import numpy as np
```

## Vector

create one Vector

```python
a = np.array([25, 10000, 5])
print(f"Vector : {a}")
print(f"wight of Vector: {len(a)}")
```

```python
# student
parsa = np.array([19, 20, 18, 16, 17, 18])
arad = np.array([18,19,16,20,15,17])
```

```python
print(f"parsa : {parsa}")
print(f"arad : {arad}")
```

```python
hasan = np.array([14.5 , 15, 16.75 , 20 , 17.25 , 18])
print(hasan)
```

```python
a = np.array([1 , 2])
b = np.array([1 , 2])
c = np.array([1 , 2])
print(a)
print(b)
print(c)
```

```python
d = e = f = np.array([3,4])
print(d)
print(e)
print(f)
```

```python
# create 3 Vector in in one line, each Vector has one unique value
g , h , i = np.array([5 ,6]) , np.array ([1 ,2]) , np.array([5 , 6])
```

```python
print(g)
print(h)
print(i)
```

```python
a = np.array([0 , 1 ,2 ,3 ,4])
print(a)

a = np.arange(5)
print(a)

a = np.arange(2, 5)
print(a)

a = np.arange(10, 1, -1) # step = -1
print(a)

a = np.arange(2, 22, 2) # step = 2
print(a)

# other way
a = np.arange(1, 11)*2 # step = 2
print(a)
```

```python
a = np.array([1,5,0,12,2,23,4])
print(a)

# indexing
print(a[0])
print(a[2])
print(a[-1])
```

Zoros and ones Vector

```python
a = np.zeros(3)
print(a)
print("--------")

a = np.zeros(7)
print(a)
print("--------")

a = np.ones(3)
print(a)
print("--------")

a = np.ones(9)
print(a)
print("--------")

a = np.ones((2,3))
print(a)
```
