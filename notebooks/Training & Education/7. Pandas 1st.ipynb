# Pandas

**pandas** is a powerful Python library for **working with tabular data and performing data analysis**.  
It provides data structures such as **DataFrame** and **Series**, similar to Excel tables or database tables, enabling fast and simple operations like reading, cleaning, filtering, aggregating, and analyzing data. It is widely used in data science, data analysis, and machine learning.

### Use Cases
- Reading and writing files (CSV, Excel, JSON, SQL)
- Data cleaning
- Filtering and selecting data
- Grouping and aggregating data (GroupBy)
- Merging and joining datasets
- Working with time series data
- Basic statistical analysis
- Preparing data for machine learning

### Oparating on data in pandas

```python
rng = np.random.RandomState(42)
ser = pd.Series(rng.randint(0, 10, 4))
ser
```

```python
df = pd.DataFrame(rng.randint(0, 10, (3,4)),
                  columns=["A", "B", "C", "D"])
df
```

```python
np.exp(ser) # use numpy in pandas
```

```python
np.sin(df *np.pi /4)
```

### Missing Data

```python
vals2 = np.array([1, "nan", 3, 4])
vals2.dtype # type nan change all type to str
```

```python
vals2 = np.array([1, np.nan, 3, 4])
vals2.dtype # nan by numpy is type of folat
```

nan safe function

```python
np.sum(vals2), np.min(vals2), np.max(vals2)
```

```python
np.nansum(vals2), np.nanmin(vals2), np.nanmax(vals2)
```

```python
vals2 = pd.Series([1, np.nan, 3, None]) # all nan is float in pandas
vals2
```

```python
vals2.isnull().isnull(), vals2.notnull()
```

```python
vals2.fillna(0), vals2.dropna()
```

```python
df = pd.DataFrame(rng.randint(0, 10, (5,4)),
                  columns=["A", "B", "C", "D"])
df.iloc[1:4, 1:3] = None
df
```

```python
df["B"] = df["B"].bfill()
df
```

```python
df["C"] = df["C"].ffill()
df
```
