# Pandas

**pandas** is a powerful Python library for **working with tabular data and performing data analysis**.  
It provides data structures such as **DataFrame** and **Series**, similar to Excel tables or database tables, enabling fast and simple operations like reading, cleaning, filtering, aggregating, and analyzing data. It is widely used in data science, data analysis, and machine learning.

### Use Cases
- Reading and writing files (CSV, Excel, JSON, SQL)
- Data cleaning
- Filtering and selecting data
- Grouping and aggregating data (GroupBy)
- Merging and joining datasets
- Working with time series data
- Basic statistical analysis
- Preparing data for machine learning

### Series

python -> list   
numpy -> array   
pandas -> Series

Series has Value and Index toghether

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
data
```

```python
data.values # this is one attribute so has NOT ()
```

```python
data.index # this is one attribute so has NOT ()
# stat index, end index, step of index
```

```python
data[1]
```

```python
data[1:3] # [1, 3)
```

```python
# change index to my index
data =pd.Series([0.25, 0.5, 0.75, 1.0],
                index = ["a", "b", "c", "d"])

data
```

```python
data["b"] # change index and b is index
```

```python
data =pd.Series([0.25, 0.5, 0.75, 1.0],
                index = [1, 12, 8, 2])
# index can be random form of number
data
```

```python
data[12]
```

```python
# Series is same Dict in python
population_dict = {"California": 3832521,
                   "Texas": 26448193,
                   "Felorida": 19552860,
                   "Illinois": 12882135}
population_dict
```

```python
population = pd.Series(population_dict)
population
```

```python
pd.Series(5, index = [100, 200, 300])
```

### DataFrame

```python
population_dict = {"California": 3832521,
                   "Texas": 26448193,
                   "Felorida": 19552860,
                   "Illinois": 12882135}
population = pd.Series(population_dict)
pd.DataFrame(population)
```

```python
pd.DataFrame({"population": population.index, "area": population})
```
