Task 3    
Data Normalization

A very common step in ML data preprocessing is **feature normalization**. This is achieved by subtracting the mean and dividing by the standard deviation for each feature.

- Calculate the **mean** and **standard deviation** of each test from the original `scores` array.
- Create a new array named `normalized_scores` using the **Z-score** formula:
  
  $$(scores - mean) / std$$

  You must correctly apply **broadcasting** to subtract the 1D `mean` and `std` arrays from the 2D `scores` matrix and then divide.
- To verify your results, calculate the mean and standard deviation for each column of `normalized_scores`. The mean should be very close to **0**, and the standard deviation should be very close to **1**.


````python
# Calculate the mean and std of each test

NUMBER_COLUMN = scores.shape[1]

for test in range(NUMBER_COLUMN):
    print(f"test number {test +1}: mean: {scores[:,test].mean():.4f} -- std: {scores[:,test].std():.4f}")
````

````python
# Create a new array named normalized_scores using the Z-score
# first way
normalized_scores = (scores - (scores.mean(axis=0))) / (scores.std(axis=0))
print(normalized_scores)
````

````python
# second way
STD_COLUMN = scores.std(axis=0) # in column
STD_MEAN = scores.mean(axis=0) # in column
normalized_scores = (scores - STD_MEAN) / STD_COLUMN
normalized_scores
````

````python
# Third way
STD_COLUMN = scores.std(axis=0) # in column
STD_MEAN = scores.mean(axis=0) # in column
normalized_scores = np.divide(np.subtract(scores, STD_MEAN), STD_COLUMN)
normalized_scores
````

````python
# test part 2 if std is 1 and mean is 0
NUMER_COLUMN_NORMALIZED = normalized_scores.shape[1]
print("std and mean nomalized score each test is by format .10f:")
for i in range(NUMER_COLUMN_NORMALIZED):
    print(f"Normalized score test number {i +1}: mean: {normalized_scores[:,i].mean():.10f} -- std: {normalized_scores[:,i].std():.10f}")
    
````
