### 1. **Data Structure Fundamentals – NumPy Basics**



Task 1   
Create a **one-dimensional NumPy array** that contains **12 consecutive integers greater than 5**.

````python
# by numpy arange
NUMBER = 5
feature_vector = np.arange(NUMBER, NUMBER + 12)
feature_vector
````

````python
# by list and numpy array
NUMBER = 5
feature_vector = np.array([i + NUMBER for i in range(12)])
feature_vector
````

````python
np.random.seed(42)
feature_matrix = np.random.randint(0, 10, size=(3, 4))
feature_matrix
````

````python
print(f"feature_vector size: {feature_vector.size}, feature_matrix size: {feature_matrix.size}")
print(f"feature_vector ndim: {feature_vector.ndim}, feature_matrix ndim: {feature_matrix.ndim}")
print(f"feature_vector shape: {feature_vector.shape}, feature_matrix shape: {feature_matrix.shape}")
print(f"feature_vector dtype: {feature_vector.dtype}, feature_matrix dtype: {feature_matrix.dtype}")
````

````python
# by dataframe and table
info_feature_vector = [feature_vector.size, feature_vector.ndim, feature_vector.shape, feature_vector.dtype]
info_feature_matrix = [feature_matrix.size, feature_matrix.ndim, feature_matrix.shape, feature_matrix.dtype]

pd.DataFrame([info_feature_vector, info_feature_matrix],
             columns=["size", "ndim", "shape", "dtype"],
             index = ["feature_vector", "feature_matrix"])
````
